```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.colors as mcolors

# Define the color palettes and convert to hex where needed
color_palettes = {
    "Tab10": [mcolors.rgb2hex(c) for c in plt.cm.tab10.colors],
    "deep": sns.color_palette("deep").as_hex(),
    "muted": sns.color_palette("muted").as_hex(),
    "pastel": sns.color_palette("pastel").as_hex(),
    "dark": sns.color_palette("dark").as_hex(),
    "colorblind": sns.color_palette("colorblind").as_hex(),
    "Set1": [mcolors.rgb2hex(c) for c in plt.cm.Set1.colors],
    "Set2": [mcolors.rgb2hex(c) for c in plt.cm.Set2.colors],
    "Set3": [mcolors.rgb2hex(c) for c in plt.cm.Set3.colors],
    "Paired": [mcolors.rgb2hex(c) for c in plt.cm.Paired.colors],
    "Classic 10: New": ["#4E79A7", "#F28E2B", "#E15759", "#76B7B2", "#59A14F", 
                        "#EDC949", "#AF7AA1", "#FF9DA7", "#9C755F", "#BAB0AC"]
}

# Create a figure with subplots for comparison
fig, axes = plt.subplots(len(color_palettes), 1, figsize=(10, len(color_palettes) * 1.5))

# Plot each palette using blocks with hex codes
for ax, (name, colors) in zip(axes, color_palettes.items()):
    ax.set_xlim(0, len(colors))
    ax.set_ylim(0, 1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(name, fontsize=12, fontweight='bold')
    
    for i, color in enumerate(colors):
        ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=color))
        ax.text(i + 0.5, 0.5, color, ha='center', va='center', fontsize=8, color='white' if np.mean(mcolors.hex2color(color)) < 0.5 else 'black')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

## Plots Work Colours

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# Define the color palettes
color_palettes = {
    "Work": ["#004077", "#c41424", "#1d1d1b", "#f4f4f4", "#7ec3ff"]
}

# Sample data for the bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))  # Ensure values match categories

# Create a figure with only as many subplots as needed
fig, axes = plt.subplots(1, len(color_palettes), figsize=(8 * len(color_palettes), 6))

# Ensure `axes` is iterable
axes = np.atleast_1d(axes)

# Plot each palette using a bar chart
for ax, (name, colors) in zip(axes, color_palettes.items()):
    ax.bar(categories, values, color=colors[:len(categories)])  # Match colors to categories
    ax.set_title(name, fontsize=14, fontweight="bold")
    ax.set_ylim(0, 120)  # Consistent y-axis

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# Define the color palettes and convert to hex where needed
color_palettes = {
    "Work": ["#004077", "#c41424", "#1d1d1b", "#f4f4f4", "#7ec3ff"]
}

# Create a figure with subplots for comparison
fig, axes = plt.subplots(len(color_palettes), 1, figsize=(10, len(color_palettes) * 1.5))

# Ensure axes is always iterable
axes = np.atleast_1d(axes)

# Plot each palette using blocks with hex codes
for ax, (name, colors) in zip(axes, color_palettes.items()):
    ax.set_xlim(0, len(colors))
    ax.set_ylim(0, 1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(name, fontsize=12, fontweight='bold')
    
    for i, color in enumerate(colors):
        ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=color))
        brightness = np.mean(mcolors.hex2color(color))  # Calculate brightness for text color
        text_color = 'white' if brightness < 0.5 else 'black'
        ax.text(i + 0.5, 0.5, color, ha='center', va='center', fontsize=8, color=text_color)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

## Work Hex Codes

* Standard Colours: #004077, #c41424
* Other Assets Colours: #1d1d1b, #f4f4f4, #7ec3ff
